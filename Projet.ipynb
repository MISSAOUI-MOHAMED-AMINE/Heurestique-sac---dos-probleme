```python
import pandas as pd
import time 
import itertools
import random 
import os
import zipfile
```

```python
df1 = pd.DataFrame(columns=[
                  'filename', 'sac capacity', "valeur_total" ,'combination capacity','duree'])
df2 = pd.DataFrame(columns=[
                  'filename', 'sac capacity', "valeur_total" ,'combination capacity','duree'])
df3 =  pd.DataFrame(columns=[
                  'filename', 'sac capacity', "valeur_total" ,'combination capacity','duree'])
df4 =  pd.DataFrame(columns=[
                  'filename', 'sac capacity', "valeur_total" ,'combination capacity','duree'])

```

```python
def all_filenames():
    file_path = 'files\\all.txt'
    try:
        with open(file_path, 'r') as f:
            l = []
            n_max = int(f.readline())
            for i in range(n_max):
                line = f.readline().strip()
                line="files\\"+line
                l.append(line.strip())
            return l
    except PermissionError as e:
        print(f"Erreur de permission : {e}")
        return None 
```

```python
def file_to_list(filename):
    l = []
    with open(filename,'r') as f :
        n_max = int(f.readline())
        c_max = int(f.readline())
        for i in range(n_max):
            line = f.readline().strip()
            id, c = line.split()
            id = int(id)
            c = int(c)
            l.append([id,c])
    return(c_max,l)
```

```python
def sort_by_ratio(elems) : 
    d = []
    ratio = []
    for i in elems : 
        ratio.append([i, i[1]/i[0]])

    ratio.sort(key=lambda x: x[1] )
    for i in ratio : 
        d.append(i[0])
    return (d)
```

```python
def find_solution_initiale(cap,elems):
    l_file = []
    solution = []
    c_max_file = 0
    for i in sort_by_ratio(elems):
        if c_max_file + i[1] <= cap:            
                l_file.append(i[0])
                c_max_file += i[1]
                solution.append(i )
    valeur_total = sum(l_file)
    return(l_file,c_max_file , cap , solution , valeur_total  )
```

```python
def generate_voisinage(element ,elems, longueur) : 
    indice = elems.index(element)
    liste_left = []
    liste_right = []
    for i in range(1,longueur+1) : 
        if (indice - i>=0 ) : 
            liste_left.append(elems[indice-i])
        if (indice+i<= len(elems)) : 
            liste_right.append(elems[indice+i])
    return (liste_left+liste_right)
```

```python
def find_solution_voisinage(cap,elems,longueur):
    l_file,c_max_file , cap  , solution , valeur_total= find_solution_initiale(cap,elems)
    amelioration = True 
    element = solution[0]
    l_file = []
    solution = []
    l_file.append(element[0])
    solution.append(element)
    c_max_file =0
    while (amelioration):
        amelioration = False
        voisinage  = generate_voisinage(element , elems , longueur)
        for i in voisinage:
            if i[0] >= element[0] and amelioration == False and c_max_file+i[1]<cap:
                element = i
                amelioration = True
                c_max_file= c_max_file+element[1]
                l_file.append(element[0])
                solution.append(element)
                break
    valeur_total = sum(l_file)
    return (l_file, c_max_file , cap , solution , valeur_total)
```

```python
def find_solution_randomized(cap,elems):
    l_file = []
    solution = []
    c_max_file = 0 
    elems_shuffled = elems.copy()
    random.shuffle(elems_shuffled)
    for i in (elems_shuffled):
        if c_max_file + i[1]  <= cap:
                l_file.append(i[0])
                c_max_file += i[1]
                solution.append(i)
    valeur_total = sum(l_file)
    return(l_file,c_max_file , cap , solution, valeur_total)
```

```python
def GRASP(cap,elems ,longueur):
    l_file,c_max_file , cap , solution, valeur_total = find_solution_randomized(cap,elems)
    amelioration = True 
    element = solution[0]
    l_file = []
    solution = []
    l_file.append(element[0])
    solution.append(element)
    c_max_file =0
    while (amelioration):
        amelioration = False
        voisinage  = generate_voisinage(element , elems , longueur)
        for i in voisinage:
            if i[0] >= element[0] and amelioration == False and c_max_file+i[1]<cap:
                element = i
                amelioration = True
                c_max_file= c_max_file+element[1]
                l_file.append(element[0])
                solution.append(element)
                break
    valeur_total = sum(l_file)
    return (l_file, c_max_file , cap , solution , valeur_total)
```

```python
def create_combination(f,cap ,elems,c_max_file,duree ,valeur_total): #file = files[i]
    a = {}
    a['filename'] = f
    a['sac capacity'] = cap
    a["valeur_total"] =  valeur_total
    a['combination capacity'] = c_max_file
    a['duree'] = duree
    return(a)
```

```python
for f in all_filenames() : 
    cap, elems = file_to_list(f)
    #########################
    t1 = time.time()
    l_file, c_max_file , cap ,solution , valeur_total= find_solution_initiale(cap, elems)
    duree = time.time()-t1
    a = create_combination(f, cap, elems, c_max_file, duree ,valeur_total)
    df1.loc[len(df1)]=a
    #########################
    t2 = time.time()
    l_file, c_max_file , cap ,solution , valeur_total = find_solution_voisinage(cap, elems ,5 )
    duree = time.time()-t2
    c = create_combination(f, cap, elems, c_max_file, duree ,valeur_total)
    df2.loc[len(df2)]=c

    ########################
    t3 = time.time()
    l_file, c_max_file , cap ,solution, valeur_total= find_solution_randomized(cap, elems)
    duree3 = time.time()-t3
    b = create_combination(f, cap,elems ,c_max_file, duree , valeur_total)
    df3.loc[len(df3)]=b
    ##########################
    t3 = time.time()
    l_file, c_max_file , cap ,solution, valeur_total= GRASP(cap, elems ,5)
    duree3 = time.time()-t3
    d = create_combination(f, cap,elems ,c_max_file, duree , valeur_total)
    df4.loc[len(df4)]=d

```
